```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme()
```

```python
df = pd.read_csv("../inputs/flow.csv")
df
```

```python
# Dividing TNFa jurkats sum by 3 because we injected 3x the cells for the single cell pop run
tnf_jurkats = df["Name"] == "Jurkat TNFa varying on rates"
df.loc[tnf_jurkats,"Aggregate(SUM)"] = df[tnf_jurkats]["Aggregate(SUM)"]//3
df
```

```python
varying_on = ["Jurkat TNFa varying on rates","varying on JTIC16"] 
varying_df = df[df["Name"].isin(varying_on)]
plt.figure(figsize=(10, 6))
plt.title("Effects of Varying Shear Stress on Adhesion")
plt.ylabel("Cells Bound (sum of positions)")
plt.xlabel("Flow Rate [ul/min]")
sns.scatterplot(data=varying_df, x="Flow", y="Aggregate(SUM)", hue="Construct",s=100)
```

```python
plt.figure(figsize=(10, 6))
plt.title("Effects of Varying Shear Stress on Adhesion")

plt.xlim([90,210])
for con in varying_df.groupby("Construct"):
    sns.regplot(x="Flow", y="Aggregate(SUM)", data=con[1], label=con[0])
plt.ylabel("Cells Bound (sum of positions)")
plt.xlabel("Flow Rate [ul/min]")
plt.legend()
```

```python
lag_series = ["ICAM lag series ON"]
lag_df = df[df["Name"].isin(lag_series)]

plt.figure(figsize=(10, 6))
plt.title("Effects of Lag Binder Strength on Adhesion (Flow Rate: 100ul/min)")
# Trying to put the Kd for the lag series but this figure isn't updating
sns.swarmplot(data=lag_df, x="Lag Binder", y="Aggregate(SUM)", hue="Construct",
              order = ['L16','L17','L42',"L18"],s=10)
plt.ylabel("Cells Bound (sum of positions)")
```
