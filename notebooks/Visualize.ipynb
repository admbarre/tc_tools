# Shear Flow Assay Visualization

* Currently this notebook is just the visualization for a set of data used for a lab meeting
* Need to standardize this into a useable pipeline for future data from shear flow assays

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme()
# This is useful!
# Poster context seems to give the thickest lines and points
sns.set_context("poster")
```

```python
df = pd.read_csv("../inputs/flow.csv")
df
```

# Appends Lag Affinities for each Lag Binder

```python
lag_affinities = {'none':0,
                  "L16": 0.7, 
                  "L17": 50, 
                  "L42": 600, 
                  "L18": 3800}
df['Affinity']=[lag_affinities[x] for x in df['Lag Binder'].values]
df
```

# Correction for Single Cell Flow Run 
1. We injected same total cells but only cell type so we have 3x the normal ammount. This might not be the standard we do later but currently dividing the sum by 3 to match with multiplexed runs

```python
# Dividing TNFa jurkats sum by 3 because we injected 3x the cells for the single cell pop run
tnf_jurkats = df["Name"] == "Jurkat TNFa varying on rates"
df.loc[tnf_jurkats,"Aggregate(SUM)"] = df[tnf_jurkats]["Aggregate(SUM)"]//3
df
```

# Subsetting DFs
1. In the future we want to do this live in a dashboard so we don't have to subset

```python
varying_on = ["Jurkat TNFa varying on rates","varying on JTIC16"] 
varying_df = df[df["Name"].isin(varying_on)]
varying_df
```

```python
lag_series = ["ICAM lag series ON"]
lag_df = df[df["Name"].isin(lag_series)]
lag_df
```

```python
con_df = varying_df[varying_df["Lag Binder"]=="none"]

con_fig, con_ax = plt.subplots(figsize=(10, 10))
con_fig.suptitle("Effects of Varying Flow Strength on Adhesion")

con_ax = sns.lineplot(
    data=con_df, x="Flow", y="Aggregate(SUM)", hue="Construct",
    palette="deep",markers=True, style="Construct", dashes=False,err_style="bars", ci=68, err_kws={'capsize':6}
)

# Adding some extras on the ends to be able to see points
con_ax.set_ylim([-10,200])
con_ax.set_ylabel("Cells Bound (sum of positions)")

con_ax.set_xlim([90,210])
con_ax.set_xlabel("Flow rates [ul/min]")
con_ax.legend(loc="upper right", bbox_to_anchor=(1.5,1))
```

# Varying Flow ON rates figure

```python
vary_fig, vary_ax = plt.subplots(figsize=(10, 10))
vary_fig.suptitle("Effects of Varying Flow Strength on Adhesion")

vary_ax = sns.lineplot(
    data=varying_df, x="Flow", y="Aggregate(SUM)", hue="Construct",
    palette="deep",markers=True, style="Construct", dashes=False,err_style="bars", ci=68, err_kws={'capsize':6}
)

# Adding some extras on the ends to be able to see points
vary_ax.set_ylim([-10,200])
vary_ax.set_ylabel("Cells Bound (sum of positions)")

vary_ax.set_xlim([90,210])
vary_ax.set_xlabel("Flow rates [ul/min]")
vary_ax.legend(loc="upper right", bbox_to_anchor=(1.5,1))
```

# Lag Series ON Rates

```python
lag_fig,lag_ax = plt.subplots(figsize=(10,10))
lag_fig.suptitle("Effects of Lag Binder Strength on Adhesion (Flow Rate: 100ul/min)")
lag_ax = sns.lineplot(
    data=lag_df, x="Affinity", y="Aggregate(SUM)", hue="ICD",
    markers=True, style="ICD", err_style="bars", ci=68, err_kws={'capsize':6}
)

# Adding some extras on the ends to be able to see points
#lag_ax.set_ylim([0,200])
lag_ax.set_ylabel("Cells Bound (sum of positions)")

lag_ax.set_xscale("log")
lag_ax.set_xlim([10**-0.5,10**4])
lag_ax.set_xlabel("Affinity: Kd [nM]")

lag_ax.legend(loc="upper right", bbox_to_anchor=(1.5,1))
```
