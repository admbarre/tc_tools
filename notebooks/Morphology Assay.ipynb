```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
directory = "/Users/adrian/code/lab/tc_tools/inputs/csv"
filename = "morpho.csv"
data = pd.read_csv(f"{directory}/{filename}")
data.head()
```

```python
measurements = ["Area", "Perim."]
fig,axes = plt.subplots(1,2, figsize=(15,5))
fig.suptitle("Morphology of NCAM vs Tether")
sns.set_theme(style="whitegrid")
for i,ax in enumerate(axes):
    sns.swarmplot(x="SynCAM",y=measurements[i], data=data, order=["Tether", "NCAM"],
                  hue="GFP", hue_order=["noGFP", "GFP"], palette=["blue", "green"], 
                  size=8, linewidth=1, dodge=True, ax=axes[i])
```
